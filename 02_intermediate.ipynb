```python
#01-Control Flow for Console
import random

print("Welcome to the High-Low Game!")
print("******************************")

total_rounds = 5

def main():
    your_score = 0

    for i in range(total_rounds):
        print("Round", i + 1)

        computer_num = random.randint(1, 10)
        your_num = random.randint(1, 10)
        print("Your number is:", your_num)

        choice = input("Do you think your number is higher or lower than the computer's? (Higher/Lower): ")

        higher_and_correct = choice.lower() == "higher" and your_num > computer_num
        lower_and_correct = choice.lower() == "lower" and your_num < computer_num

        if higher_and_correct or lower_and_correct:
            print("You are right! The computer's number was:", computer_num)
            your_score += 1
        else:
            print("That's incorrect. The computer's number was:", computer_num)

        print("Your score is now:", your_score)
        print()

    print("Thanks for playing! Your final score is:", your_score)

if __name__ == '__main__':
    main()
```

```python
#02- Intro to Python
def main():
    print("******** WELCOME TO PLANETARY WEIGHT CALCULATOR **********")
    earth_weight = float(input("Enter your weight on earth: "))
    gravity_ratio = {
        "Mercury": 0.38,
        "Venus": 0.91,
        "Mars": 0.38,
        "Jupiter": 2.34,
        "Saturn": 1.06,
        "Uranus": 0.92,
        "Neptune": 1.19
    }

    print("\nSelect a planet:")
    for planet in gravity_ratio:
        print(planet)

    planet_choice = input("\nEnter your choice: ").title()

    if planet_choice in gravity_ratio:
        planet_weight = earth_weight * gravity_ratio[planet_choice]
        print(f"\nYour weight on {planet_choice} is {planet_weight:.2f} kg.")
    else:
        print("\nInvalid planet choice.")

if __name__ == '__main__':
    main()

```

```python
# 03 - Lists and Dicts

my_list = ["apple", "mango", "banana", "orange", "grapes"]

def access_element(my_list, index):
    """Accesses an element in a list at a given index."""
    if 0 <= index < len(my_list):
        return f'Element at index {index}: {my_list[index]}'
    return "Index out of range"

def modify_element(my_list, index, new_value):
    """Modifies an element in a list at a given index with a new value."""
    if 0 <= index < len(my_list):
        my_list[index] = new_value
        return f'Element at index {index} modified to {new_value}'
    return "Index out of range"

def slice_list(my_list, start, end):
    """Slices a list from start index (inclusive) to end index (exclusive)."""
    if 0 <= start < len(my_list) and 0 <= end <= len(my_list) and start < end:
        return f'Sliced list: {my_list[start:end]}'
    return "Invalid slice indices"

def list_game():
    print("\nðŸŽ® Welcome to the List Manipulation Game!")
    my_list = ["apple", "mango", "banana", "orange", "grapes"]

    while True:
        print("\nCurrent List:", my_list)
        print("\nChoose an option:")
        print("1. Access element")
        print("2. Modify element")
        print("3. Slice list")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            index = int(input("Enter index to access: "))
            print(access_element(my_list, index))
        elif choice == "2":
            index = int(input("Enter index to modify: "))
            new_value = input("Enter new value: ")
            print(modify_element(my_list, index, new_value))
        elif choice == "3":
            start = int(input("Enter start index: "))
            end = int(input("Enter end index: "))
            print(slice_list(my_list, start, end))
        elif choice == "4":
            print("Exiting the game. Bye! ðŸ‘‹")
            break
        else:
            print("Invalid choice. Please enter a number from 1 to 4.")

if __name__ == '__main__':
    list_game()



```
